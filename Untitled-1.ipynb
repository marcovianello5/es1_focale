```julia
# Tutte le misure sono in mm

# Le misure in x, y, z, sono riferite agli oggetti disposti lungo l'asse ottico z in posizione di lavoro

#     x
#     |
#     |
#     |_ _ _a_s_s_e_ _o_t_t_i_c_o_ _ _ _ z
#    /
#   /
#  y

# La scritta THORLABS di Lente₁ deve guardare verso sinistra

schermoz = 3  # Da CAD. Dimensione in z dello schermo
schermoy = 37.5 * 2  # Da CAD. Dimensione in y dello schermo

carrelloz = 12.7 * 2  # Da CAD. Dimensione in z della base

bacchettaz = 81  # Misurato da noi. Dimensione in z della bacchetta

# Specifiche della lente
vv_moodle = 11.8  # Da Moodle
R = 75.2  # Da Moodle

#zoggetto₁ = 39  # Misurato da noi. Riferirsi all'appendice. Misure del secondo giorno
z = 43  # Misurato da noi. Riferirsi all'appendice. Misure dell'ultimo giorno

δ1 = bacchettaz + 1/2*schermoz - 1/2*carrelloz - z  # Offset oggetto-inizio della guida metrica
@show δ1

δ2 = bacchettaz + 1/2*schermoz - 1/2*carrelloz - (167 - 100)  # Offset lato sinistro del carrello-vertice sinistro della lente
δ3 = bacchettaz + 1/2*schermoz - 1/2*carrelloz - (253 - (167 + carrelloz))  # Offset vertice destro della lente-lato destro del carrello
@show δ2, δ3

δ4 = δ2 + 1/2*vv_moodle - 1/2*(vv_moodle/3)  # Offset lato sinistro del carrello-piano principale sinistro della lente
δ5 = δ3 + 1/2*vv_moodle - 1/2*(vv_moodle/3)  # Offset piano principale sinistro della lente-lato destro del carrell
@show δ4, δ5

δ6 =  1/2*carrelloz + 3 # Offset lato sinistro del carrello-sensore CCD
@show δ6
```

### Metodo dei punti coniugati

```julia
punticoniugati_df = CSV.read("dati/punticoniugati.csv", DataFrame, delim=" ")
delete!(punticoniugati_df, [2, 4, 6, 8, 10])
```

```julia
# Applico le correzioni nell'ipotesi di lente sottile

#punticoniugati_df.p = @. δoggetto(zoggetto₁) + punticoniugati_df.lente + δlente_Π_sx(z1lente1, z2lente1, vv_misurato1)
#punticoniugati_df.q = @. δoggetto(zoggetto₁) + punticoniugati_df.camera + δcamera - vv_misurato1/3 - punticoniugati_df.p

punticoniugati_df.p = @. δ1 + punticoniugati_df.lente + δ4
punticoniugati_df.q = @. punticoniugati_df.camera - punticoniugati_df.lente - carrelloz + δ5 + δ6

punticoniugati_df.X = @. 1 / punticoniugati_df.p
punticoniugati_df.Y = @. 1 / punticoniugati_df.q

punticoniugati_df.f = @. (punticoniugati_df.X + punticoniugati_df.Y)^(-1)
```

```julia
CSV.write("typst/dati/punticoniugati_df_sample.csv", round.(punticoniugati_df[1:10:end, [:lente, :camera, :p, :q]], digits=4), header=false)
```

```julia
# Plotto le misure grezze

pq = scatter(punticoniugati_df.p, punticoniugati_df.q, ms=3, mc=:black, msw=0, label="")
xlabel!(L"p\quad [\mathrm{mm}]")
ylabel!(L"q\quad [\mathrm{mm}]")

XY = scatter(punticoniugati_df.X, punticoniugati_df.Y, ms=3, mc=:black, msw=0, label="")
xlabel!(L"1/p\quad [1/\mathrm{mm}]")
ylabel!(L"1/q\quad [1/\mathrm{mm}]")

savefig(pq, "typst/grafici/pq.svg")
savefig(XY, "typst/grafici/XY.svg")
```

```julia
#Plotto X vs. Y

p = scatter(punticoniugati_df.X, punticoniugati_df.Y, label="dati", markersize=3)
```

```julia
# Minimi quadrati sulle variabili X e Y

n = length(punticoniugati_df.X)
A = [ones(n) punticoniugati_df.X]

b0, b1 = A \ punticoniugati_df.Y

f1 = b0^(-1)
print("lunghezza focale f₁: $(f1)")
```

```julia
# Plotto la retta di regressione determinata sopra

plot(p,
    punticoniugati_df.X,
    (@. b0 + b1 * punticoniugati_df.X),
    label="fit lineare",
    linewidth=2,
    xlabel="1/p",
    ylabel="1/q")
```

```julia
# Provo a limitare la regione dove fittare

filtro = punticoniugati_df.lente .>= 200

filtrato = punticoniugati_df[filtro, :];
```

```julia
# Minimi quadrati su una porzione dei dati su regione limitata


n = length(filtrato.X)
A = [ones(n) filtrato.X]

b0, b1 = A \ filtrato.Y

f1 = b0^(-1)
print("lunghezza focale f₁: $(f1)")
```

```julia
# Plotto la nuova retta di regressione

plot(p,
    filtrato.X,
    (@. b0 + b1 * filtrato.X),
    label="fit lineare su\nregione limitata",
    linewidth=2,
    xlabel="1/p",
    ylabel="1/q")
```

```julia
# Minimi quadrati con GLM.jl

model = lm(@formula(Y ~ X), punticoniugati_df)

result = coef(model)

f1 = result[1]^(-1)
```

```julia
# Plotto la retta di regressione determinata con GLM.jl

plot(p,
    punticoniugati_df.X,
    (@. result[1] + result[2] * punticoniugati_df.X),
    label="fit lineare",
    linewidth=2,
    xlabel="1/p",
    ylabel="1/q")
```

```julia
# Minimi quadrati con LsqFit su regione limitata e coefficiente angolare fissato a -1

using LsqFit

modello(x, p) = @. p[1] + (-1)*x

p0 = [1/80]

fit = curve_fit(modello, filtrato.X, filtrato.Y, p0)

f1 = fit.param[1]^(-1)
print("lunghezza focale f₁: $(f1)")
```

```julia
# Plotto la roba sopra

plot(p,
    filtrato.X,
    (@. fit.param[1] - 1 * filtrato.X),
    label="fit lineare a coeff. ang. -1\nsu regione limitata",
    linewidth=2,
    xlabel="1/p",
    ylabel="1/q")
```
