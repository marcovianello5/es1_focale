## Offset

```julia
# Tutte le misure sono in mm

# Le misure in x, y, z, sono riferite agli oggetti disposti lungo l'asse ottico z in posizione di lavoro

#     x
#     |
#     |
#     |_ _ _ z
#    /
#   /
#  y

# La scritta THORLABS di Lente₁ deve guardare verso sinistra

schermoz = 3  # Da CAD. Dimensione in z dello schermo
schermoy = 37.5 * 2  # Da CAD. Dimensione in y dello schermo

basez = 12.7 * 2  # Da CAD. Dimensione in z della base

bacchettaz = 81  # Misurato da noi. Dimensione in z della bacchetta

# Specifiche della lente
vv_moodle = 11.8  # Da Moodle
R = 75.2  # Da Moodle

zoggetto₁ = 39  # Misurato da noi. Riferirsi all'appendice. Misure del secondo giorno
zoggetto₂ = 43  # Misurato da noi. Riferirsi all'appendice. Misure dell'ultimo giorno

# Offset oggetto-lato sinistro della base
# Dipende dal valore z misurato a sinistra del carrello
function δoggetto(z)
    return bacchettaz + schermoz/2 - z - basez/2
end

z1lente1 = 100  # Misurato da noi. Riferirsi all'appendice
z2lente1 = 167  # Misurato da noi. Riferirsi all'appendice
z3lente1 = 253  # Misurato da noi. Riferirsi all'appendice

z1lente2 = 0  # Mi sono dimenticato di misurarlo prkd
z2lente2 = 0  # Same
z3lente2 = 0  # Same

# Offset lato sinistro della base-vertice sinistro di Lente₁
# Dipende dai valori z1 e z2 definiti nell'appendice
function δlente_V_sx(z1, z2)
    Δ = z2 - z1

    return schermoz/2 + bacchettaz - basez/2 - Δ
end

# Offset vertice destro della lente-vertice destro della base
# Dipende dai valori z1 e z2 definiti nell'appendice
function δlente_V_dx(z2, z3)
    Δ = z3 - (z2 + basez)

    return bacchettaz + schermoz/2 - basez/2 - Δ
end

vv_misurato1 = basez - (δlente_V_sx(z1lente1, z2lente1) + δlente_V_dx(z2lente1, z3lente1))

# Offset lato sinistro della base-piano principale sinistro di Lenteᵢ per i=1,2
# Dipende dai valori z1 e z2 definiti nell'appendice e dalla distanza vertice-vertice di Lenteᵢ
function δlente_Π_sx(z1, z2, vv)
    return δlente_V_sx(z1, z2) + vv/2 - (vv/3)/2
end

# Offset piano principale destro di Lenteᵢ-lato destro della base per i=1,2
# Dipende dai valori z1 e z2 definiti nell'appendice e dalla distanza vertice-vertice di Lenteᵢ
function δlente_Π_dx(z2, z3, vv)
    return vv/2 + δlente_V_dx(z2, z3) - (vv/3)/2
end

cameraz_asse = 6  # Da CAD. Distanza fronte camera-asse camera
cameraz_ip = 9  # Da CAD. Distanza fronte camera-sensore (piano immagine)

δcamera = basez/2 + (cameraz_ip - cameraz_asse)  # Offset lato sinistro della base-sensore (piano immagine)
```

## Stima della lunghezza focale $ f_1 $ di Lente₁

### Metodo dei punti coniugati

```julia
punticoniugati_df = CSV.read("punticoniugati.csv", DataFrame, delim=" ")
```

```julia
# Plotto le misure grezze

scatter(punticoniugati_df.lente, punticoniugati_df.camera, label="")
xlabel!("lente")
ylabel!("camera")
```

```julia
# Applico le correzioni nell'ipotesi di lente sottile

punticoniugati_df.p = @. δoggetto(zoggetto₁) + punticoniugati_df.lente + δlente_Π_sx(z1lente1, z2lente1, vv_misurato1)
punticoniugati_df.q = @. δoggetto(zoggetto₁) + punticoniugati_df.camera + δcamera - vv_misurato1/3 - punticoniugati_df.p

punticoniugati_df.X = @. 1 / punticoniugati_df.p
punticoniugati_df.Y = @. 1 / punticoniugati_df.q

punticoniugati_df.f = @. (punticoniugati_df.X + punticoniugati_df.Y)^(-1)
```

```julia
# Plotto la focale fᵢ corrispondente alla coppia (pᵢ, qᵢ) per i=1,...

plot(punticoniugati_df.f, label="f di Lente₁ (punti coniugati)", size=(800,600), xticks=1:2:length(punticoniugati_df.f))
plot!(punticoniugati_df[punticoniugati_df.lente .>=200, :].f)
xlabel!("# misura")
ylabel!("f")
```

```julia
Plotto X vs. Y

p = scatter(punticoniugati_df.X, punticoniugati_df.Y, label="dati", markersize=3)
```

```julia
# Minimi quadrati sulle variabili X e Y

n = length(punticoniugati_df.X)
A = [ones(n) punticoniugati_df.X]

b0, b1 = A \ punticoniugati_df.Y

f1 = b0^(-1)
print("lunghezza focale f₁: $(f1)")
```

```julia
# Plotto la retta di regressione determinata sopra

plot(p,
    punticoniugati_df.X,
    (@. b0 + b1 * punticoniugati_df.X),
    label="fit lineare",
    linewidth=2,
    xlabel="1/p",
    ylabel="1/q")
```

```julia
# Provo a limitare la regione dove fittare

filtro = punticoniugati_df.lente .>= 200

filtrato = punticoniugati_df[filtro, :];
```

```julia
# Minimi quadrati su una porzione dei dati su regione limitata


n = length(filtrato.X)
A = [ones(n) filtrato.X]

b0, b1 = A \ filtrato.Y

f1 = b0^(-1)
print("lunghezza focale f₁: $(f1)")
```

```julia
# Plotto la nuova retta di regressione

plot(p,
    filtrato.X,
    (@. b0 + b1 * filtrato.X),
    label="fit lineare su\nregione limitata",
    linewidth=2,
    xlabel="1/p",
    ylabel="1/q")
```

```julia
# Minimi quadrati con GLM.jl

model = lm(@formula(Y ~ X), punticoniugati_df)

result = coef(model)

f1 = result[1]^(-1)
```

```julia
# Minimi quadrati con LsqFit su regione limitata e coefficiente angolare fissato a -1

using LsqFit

modello(x, p) = @. p[1] + (-1)*x

p0 = [1/80]

fit = curve_fit(modello, filtrato.X, filtrato.Y, p0)

f1 = fit.param[1]^(-1)
print("lunghezza focale f₁: $(f1)")
```

```julia
# Plotto la roba sopra

plot(p,
    filtrato.X,
    (@. fit.param[1] - 1 * filtrato.X),
    label="fit lineare a coeff. ang. -1\nsu regione limitata",
    linewidth=2,
    xlabel="1/p",
    ylabel="1/q")
```
