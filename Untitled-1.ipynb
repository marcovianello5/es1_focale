```julia
# Faccio i fit

@. nonlineare(p, parametri) = (parametri[1] - 1/p)^(-1)
@. lineare(X, parametri) = parametri[1] - parametri[2] * X

p, q = Measurements.value.(punticoniugati_df.p), Measurements.value.(punticoniugati_df.q)
X, Y = Measurements.value.(punticoniugati_df.X), Measurements.value.(punticoniugati_df.Y)

q_err = 1 ./ (Measurements.uncertainty.(punticoniugati_df.q)) .^ 2
Y_err = 1 ./ (Measurements.uncertainty.(punticoniugati_df.Y)) .^ 2

fit_nonlineare = curve_fit(nonlineare, p, q, q_err, [1/80])
fit_lineare = curve_fit(lineare, X, Y, Y_err, [1/80, -1])

@show fit_nonlineare.param[1]
@show fit_lineare.param[1]
```

## Stima delle incertezze

```julia
# Funzione gaussiana
@. model(x, p) = p[1] * exp(-0.5 * ((x - p[2]) / p[3])^2) + p[4]

poss = [0.0, 0.5, -0.5, 1.0, -1.0]  # in mm
files = ["img/stack/43_52-5_500_000$(i).csv" for i in 1:5]

results = DataFrame(
    position = Float64[],
    sigma = Float64[],
    sigma_err = Float64[],
    fwhm = Float64[],
    fwhm_err = Float64[],
    amplitude = Float64[],
    center = Float64[],
    background = Float64[],
    R2 = Float64[]
)

for (pos, file) in zip(poss, files)
    df = CSV.read(file, DataFrame, delim=",")
    x = df.distance
    y = df.intensity
    
    A_guess = maximum(y) - minimum(y)
    μ_guess = x[argmax(y)]
    σ_guess = 5.0  # Guess tipico per un punto focale
    B_guess = minimum(y)
    p0 = [A_guess, μ_guess, σ_guess, B_guess]
    
    fit = curve_fit(model, x, y, p0)
    
    params = coef(fit)
    errors = stderror(fit)
    
    # Calcola FWHM e suo errore
    σ = params[3]
    σ_err = errors[3]
    fwhm = 2.355 * σ
    fwhm_err = 2.355 * σ_err
    
    # Calcola R²
    y_pred = model(x, params)
    SS_res = sum((y .- y_pred).^2)
    SS_tot = sum((y .- mean(y)).^2)
    R2 = 1 - SS_res/SS_tot
    
    # Salva risultati
    push!(results, (pos, σ, σ_err, fwhm, fwhm_err, params[1], params[2], params[4], R2))
end

println(results)

ppos_fwhm = scatter(results.position, results.fwhm, yerror=results.fwhm_err, label="Dati", xlabel="Posizione camera [mm] rispetto a q₀", ylabel="FWHM [pixel]", title="Nitidezza vs Posizione")

fwhm_min = minimum(results.fwhm)
threshold = fwhm_min * 1.20  # 20% di tolleranza

# Linea della soglia
hline!(ppos_fwhm, [threshold], linestyle=:dash, color=:red, label="Soglia 20% = $(round(threshold, digits=1)) pixel")

# Fit parabolico per interpolare (opzionale, ma utile)
x_fine = range(-1.0, 1.0, length=100)
para_fit = curve_fit((x, p) -> p[1]*x.^2 .+ p[2]*x .+ p[3], results.position, results.fwhm, [1.0, 0.0, fwhm_min])
p_para = coef(para_fit)
plot!(x_fine, p_para[1]*x_fine.^2 .+ p_para[2]*x_fine .+ p_para[3], label="Fit parabolico", color=:green)

# Trova dove la parabola interseca la soglia
# (Risolvi p[1]*x² + p[2]*x + p[3] = threshold)
a, b, c = p_para[1], p_para[2], p_para[3] - threshold
Δq_estimate = sqrt(b^2 - 4*a*c) / (2*abs(a))  # Soluzione positiva

println("\n=== RISULTATI ===")
println("FWHM minima: $(round(fwhm_min, digits=1)) pixel")
println("Soglia (20%): $(round(threshold, digits=1)) pixel")
println("Δq stimato: ±$(round(Δq_estimate, digits=2)) mm")
println("Raccomandazione: Δq = $(round(Δq_estimate, digits=1)) mm")

plot(ppos_fwhm)
```

```julia
@. gaus(x, p) = p[1] * exp(-1/2 * ((x - p[2]) / p[3])^2) + p[4]

poss = [0.0, 0.5, -0.5, 1.0, -1.0]  # in mm
files = ["img/stack/diametro/secondo/43_52-5_500_000$(i).csv" for i in 1:5]

risultati = DataFrame(
    posizione=Float64[],
    σ=Float64[],
    σ_err=Float64[],
    fwhm=Float64[],
    fwhm_err=Float64[],
    ampiezza=Float64[],
    μ=Float64[],
    background=Float64[]
)

p = plot()

for (pos, file) in zip(poss, files)
    df = CSV.read(file, DataFrame, delim=",")
    # roi_df = df[(df.distance .>= 220) .& (df.distance .<= 1600), :]
    roi_df = df

    distance = roi_df[:, 1]
    intensity = roi_df[:, 2]

    plot!(p, distance, intensity, label=pos, xticks=0:200:1700)

    p01 = maximum(intensity) - minimum(intensity)
    p02 = distance[argmax(intensity)]
    p03 = 80  # Numero a caso finché funzia
    p04 = minimum(intensity)
    p0 = [p01, p02, p03, p04]

    @show p0

    gaus_fit = curve_fit(gaus, distance, intensity, p0)

    plot!(p, 0:1:1100, gaus.(0:1:1100, Ref(gaus_fit.param)), label="gaus $(pos)")

    ampiezza, μ, σ, background = coef(gaus_fit)
    σ_err = stderror(gaus_fit)[3]

    # Calcolo FWHM
    fwhm = 2.355 * σ
    fwhm_err = 2.355 * σ_err

    # Salvo
    push!(risultati, (pos, σ, σ_err, fwhm, fwhm_err, ampiezza, μ, background))
end

CSV.write("typst/dati/fwhm.csv", round.(risultati[1:end, [:posizione, :σ, :σ_err, :fwhm, :fwhm_err]], digits=2), header=false)

plot(p)
```
